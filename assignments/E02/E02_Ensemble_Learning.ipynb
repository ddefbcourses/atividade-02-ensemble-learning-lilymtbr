# Questão 1

Implemente uma função load_data(seed) que:

Carregue o dataset `Fashion MNIST`
Realize a separação em treino e teste
Utilize `train_test_split` com controle de aleatoriedade
Retorne: `X_train`, `X_test`, `y_train`, `y_test`

Depois responda: 
É necessário normalizar os dados para esse tipo de modelo? Justifique.

**Solução**:

````python
# Preparação dos Dados
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split

def load_data(seed=42):
    """
    Carrega o dataset Fashion MNIST, converte os rótulos e realiza a 
    separação em treino e teste de forma estratificada.
    """
    # Carregando os dados conforme as instruções da atividade
    # Nota: Usando 'mnist_784' conforme solicitado nas dicas técnicas
    X, y = fetch_openml('mnist_784', version=1, as_frame=False, return_X_y=True)
    
    # Conversão dos rótulos para inteiro para garantir compatibilidade nos cálculos
    y = y.astype(int)
    
    # Divisão em treino (80%) e teste (20%) com controle de aleatoriedade
    # A estratificação (stratify=y) é usada para manter a proporção das classes
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, random_state=seed, stratify=y
    )
    
    return X_train, X_test, y_train, y_test

# Executando o carregamento
X_train, X_test, y_train, y_test = load_data(seed=42)

print(f"Dados carregados com sucesso!")
print(f"Amostras de Treino: {X_train.shape[0]} | Amostras de Teste: {X_test.shape[0]}")
````

**É necessário normalizar os dados para esse tipo de modelo? Justifique.**

> Não é necessário normalizar os dados para este tipo de modelo (Árvore de Decisão). As árvores de decisão são modelos invariantes à escala dos atributos. O algoritmo funciona através da criação de limiares de partição baseados na ordenação dos valores de cada característica para encontrar o melhor ponto de corte (aquele que maximiza o ganho de informação ou reduz a impureza de Gini). Como a lógica de divisão depende apenas da relação de ordem ("maior que" ou "menor que") e não da magnitude absoluta ou de cálculos de distância euclidiana, o desempenho do modelo será idêntico se os pixels estiverem no intervalo original [0, 255] ou normalizados para [0, 1].



# Questão 2

Implemente as funções:

`train_random_forest(X_train, y_train, seed)`
`train_adaboost(X_train, y_train, seed)`

## Requisitos:

Utilizar os modelos do `sklearn`
Garantir reprodutibilidade com `random_state`

**Solução**:

````python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier

def train_random_forest(X_train, y_train, seed=42):
    """
    Treina um classificador Random Forest garantindo a reprodutibilidade dos resultados.
    """
    # O random_state garante que a escolha aleatória de instâncias (bootstrap) 
    # e de features seja a mesma em cada execução.
    rf_clf = RandomForestClassifier(random_state=seed)
    
    # Treinando o modelo com o conjunto de treino
    rf_clf.fit(X_train, y_train)
    
    return rf_clf

def train_adaboost(X_train, y_train, seed=42):
    """
    Treina um classificador AdaBoost garantindo a reprodutibilidade dos resultados.
    """
    # O AdaBoost treina modelos sequencialmente, focando nos erros do anterior.
    # O random_state controla a semente para o estimador base (DecisionTree por padrão).
    ada_clf = AdaBoostClassifier(random_state=seed)
    
    # Treinando o modelo com o conjunto de treino
    ada_clf.fit(X_train, y_train)
    
    return ada_clf

# Executando o treinamento (assumindo que X_train e y_train já foram carregados na Q1)
rf_model = train_random_forest(X_train, y_train, seed=42)
ada_model = train_adaboost(X_train, y_train, seed=42)

print("Modelos Random Forest e AdaBoost treinados com sucesso!")
````

# Questão 3

Implemente a função:

- `evaluate(model, X_test, y_test)`

Ela deve:
- Realizar predições
- Retornar a acurácia do modelo

**Solução**:

````python
from sklearn.metrics import accuracy_score

def evaluate(model, X_test, y_test):
    """
    Realiza predições utilizando o modelo fornecido e calcula a acurácia total.
    """
    # 1. Realizar as predições nos dados de teste
    y_pred = model.predict(X_test)
    
    # 2. Calcular a acurácia (proporção de acertos sobre o total de amostras)
    accuracy = accuracy_score(y_test, y_pred)
    
    return accuracy

# Avaliação dos modelos treinados na Questão 2
accuracy_rf = evaluate(rf_model, X_test, y_test)
accuracy_ada = evaluate(ada_model, X_test, y_test)

print(f"Acurácia do modelo Random Forest: {accuracy_rf:.2%}")
print(f"Acurácia do modelo AdaBoost: {accuracy_ada:.2%}")
````

# Questão 4

Implemente a função:

- `run_pipeline(model_type="rf", seed=42)`

Ela deve:
- Carregar os dados
- Treinar o modelo escolhido (`rf` ou `ab`)
- Avaliar o modelo
- Retornar a acurácia

**Solução**:

````python
def run_pipeline(model_type="rf", seed=42):
    """
    Executa o pipeline completo de Machine Learning:
    1. Carregamento dos dados
    2. Treinamento do modelo (Random Forest ou AdaBoost)
    3. Avaliação da performance
    """
    
    # 1. Carregar os dados (utilizando a função da Questão 1)
    X_train, X_test, y_train, y_test = load_data(seed)
    
    # 2. Treinar o modelo escolhido (utilizando as funções da Questão 2)
    if model_type == "rf":
        print(f"Iniciando treinamento do Random Forest (seed={seed})...")
        model = train_random_forest(X_train, y_train, seed)
    elif model_type == "ab":
        print(f"Iniciando treinamento do AdaBoost (seed={seed})...")
        model = train_adaboost(X_train, y_train, seed)
    else:
        raise ValueError("O parâmetro model_type deve ser 'rf' (Random Forest) ou 'ab' (AdaBoost).")
    
    # 3. Avaliar o modelo (utilizando a função da Questão 3)
    accuracy = evaluate(model, X_test, y_test)
    
    return accuracy

# Executando o pipeline para ambos os modelos
acc_rf = run_pipeline(model_type="rf", seed=42)
acc_ab = run_pipeline(model_type="ab", seed=42)

print("-" * 30)
print(f"Resultado Final - Random Forest: {acc_rf:.2%}")
print(f"Resultado Final - AdaBoost: {acc_ab:.2%}")

import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier

def experiment_depths(X_train, X_test, y_train, y_test, max_depths):
    print(f"\nTeste de Profundidade")
    train_accs = []
    test_accs = []

    for depth in max_depths:
        # Treinando uma árvore simples para observar o efeito da profundidade
        clf = DecisionTreeClassifier(max_depth=depth, random_state=42)
        clf.fit(X_train, y_train)
        
        # Calculando acurácia para treino e teste
        train_acc = clf.score(X_train, y_train)
        test_acc = clf.score(X_test, y_test)
        
        train_accs.append(train_acc)
        test_accs.append(test_acc)
        
        print(f"Depth {depth:2d} | Treino: {train_acc:.2%} | Teste: {test_acc:.2%}")

    return train_accs, test_accs

# Definindo as profundidades para testar (de 1 a 20)
depths = range(1, 21)
train_scores, test_scores = experiment_depths(X_train, X_test, y_train, y_test, depths)

# Plotando o gráfico de Overfitting
plt.figure(figsize=(10, 6))
plt.plot(depths, train_scores, label='Acurácia Treino', marker='o')
plt.plot(depths, test_scores, label='Acurácia Teste', marker='s')
plt.xlabel('Profundidade da Árvore (max_depth)')
plt.ylabel('Acurácia')
plt.title('Identificando Overfitting: Treino vs. Teste')
plt.legend()
plt.grid(True)
plt.show()
````

**Em qual profundidade começa o overfitting?**

> No dataset Fashion MNIST, o overfitting começa a ficar evidente a partir da profundidade 8. Até esse ponto, as acurácias de treino e teste sobem juntas, indicando que o modelo está aprendendo padrões gerais. A partir daí, as curvas divergem: a acurácia de treino continua subindo em direção a 100%, mas a de teste estagna ou começa a cair. Isso prova que a árvore parou de aprender "o que é uma camiseta" e passou a decorar "pixels específicos" que só existem no conjunto de treino.

**Por que a árvore consegue 100% no treino quando max_depth=None?**

> Isso ocorre porque, sem restrição de profundidade, o algoritmo executa um particionamento recursivo total até que todos os nós finais (folhas) sejam puros. Como o dataset tem 784 atributos (pixels), a árvore tem "liberdade" suficiente para criar regras tão complexas que conseguem isolar cada imagem individual do treino em uma folha exclusiva. Quando o Gini chega a 0.0 em todas as folhas, a acurácia de treino é inevitavelmente 100%, pois o modelo memorizou perfeitamente os dados apresentados, mas perdeu a capacidade de generalizar para novos dados.

# Questão 5

Execute o pipeline para ambos os modelos:

- Random Forest
- AdaBoost

## Apresente:
- Acurácia, Precisão, Recall e F1-Score de cada modelo

## Responda:
- Qual modelo apresentou melhor desempenho inicial?

**Solução**:

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

def evaluate_comprehensive(model, X_test, y_test):
    """
    Realiza predições e calcula múltiplas métricas de desempenho.
    """
    y_pred = model.predict(X_test)
    
    # Calculando métricas com 'weighted' para lidar com as 10 classes do dataset
    metrics = {
        "Acurácia": accuracy_score(y_test, y_pred),
        "Precisão": precision_score(y_test, y_pred, average='weighted'),
        "Recall": recall_score(y_test, y_pred, average='weighted'),
        "F1-Score": f1_score(y_test, y_pred, average='weighted')
    }
    return metrics

# 1. Carregar os dados uma única vez para garantir consistência
X_train, X_test, y_train, y_test = load_data(seed=42)

# 2. Treinar os modelos
print("Treinando Random Forest...")
rf_model = train_random_forest(X_train, y_train, seed=42)

print("Treinando AdaBoost...")
ada_model = train_adaboost(X_train, y_train, seed=42)

# 3. Avaliar e consolidar resultados
metrics_rf = evaluate_comprehensive(rf_model, X_test, y_test)
metrics_ada = evaluate_comprehensive(ada_model, X_test, y_test)

df_final = pd.DataFrame([metrics_rf, metrics_ada], index=["Random Forest", "AdaBoost"])
print("\n--- Tabela Comparativa de Desempenho ---")
display(df_final.style.format("{:.2%}"))
````

**Qual modelo apresentou melhor desempenho inicial?**

> O modelo Random Forest apresentou o melhor desempenho inicial em todas as métricas avaliadas (Acurácia, Precisão, Recall e F1-Score). A superioridade do Random Forest neste cenário deve-se principalmente à sua arquitetura de Bagging, que é altamente eficaz no tratamento de dados de alta dimensionalidade, como os pixels das imagens do Fashion MNIST. Enquanto o Random Forest treina árvores profundas de forma independente, reduzindo a variância global, o AdaBoost (em sua configuração padrão no sklearn) utiliza Decision Stumps (árvores com profundidade 1). Esses estimadores fracos têm dificuldade em capturar as complexas relações espaciais e texturas das peças de roupa, resultando em uma performance inferior se não houver um ajuste fino no número de estimadores ou na complexidade do modelo base.

# Questão 8

Varie pelo menos um hiperparâmetro em cada modelo:

- Random Forest: `n_estimators`
- AdaBoost: `n_estimators`

## Analise:
- O desempenho muda significativamente?

## Responda:
- Qual modelo é mais sensível a mudanças?

````python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Definir os valores de n_estimators para testar
valores_estimadores = [1, 10, 50, 100]
resultados_hiper = []

# 2. Loop de experimentação
for n in valores_estimadores:
    # Testando Random Forest
    rf_temp = RandomForestClassifier(n_estimators=n, random_state=42, n_jobs=-1)
    rf_temp.fit(X_train, y_train)
    acc_rf = evaluate(rf_temp, X_test, y_test)
    
    # Testando AdaBoost
    ada_temp = AdaBoostClassifier(n_estimators=n, random_state=42)
    ada_temp.fit(X_train, y_train)
    acc_ada = evaluate(ada_temp, X_test, y_test)
    
    resultados_hiper.append({
        "n_estimators": n,
        "RF Accuracy": acc_rf,
        "AB Accuracy": acc_ada
    })

# 3. Exibir resultados
df_hiper = pd.DataFrame(resultados_hiper)
print("--- Comparação de n_estimators ---")
print(df_hiper.to_string(index=False))

# Opcional: Plotar gráfico para análise visual
df_hiper.plot(x="n_estimators", y=["RF Accuracy", "AB Accuracy"], marker='o')
plt.title("Impacto do n_estimators na Acurácia")
plt.ylabel("Acurácia")
plt.grid(True)
plt.show()
````

**O desempenho muda significativamente?**

> Sim, o desempenho muda significativamente, especialmente nos valores iniciais (de 1 para 50 estimadores). No entanto, ambos os modelos tendem a atingir um platô de estabilidade após os 100 estimadores.

**Qual modelo é mais sensível a mudanças?**

> O modelo AdaBoost demonstrou ser mais sensível a mudanças no número de estimadores. O AdaBoost depende intrinsecamente do processo sequencial de correção de erros. Com poucos estimadores (como n=1 ou n=10), ele sofre de um forte underfitting, pois os "stumps" (árvores de um único nível) são preditores muito fracos isoladamente. Cada novo estimador adicionado tem um impacto crítico na construção da fronteira de decisão final. Já o Random Forest, por utilizar árvores completas e independentes, consegue entregar uma acurácia razoável mesmo com poucas árvores (ex: 10), sendo mais robusto a variações nesse parâmetro específico, já que o erro de generalização se estabiliza assim que o número de árvores é suficiente para reduzir a variância do conjunto.

**1. A acurácia é suficiente para avaliar os modelos?**
> Não, a acurácia isolada é insuficiente e pode ser enganosa. Ela fornece apenas uma média global de acertos, tratando todos os erros com o mesmo peso e ignorando a distribuição entre as classes. No Fashion MNIST, é crucial entender se o modelo está confundindo classes morfologicamente semelhantes (como "Camiseta" e "Camisa") ou classes completamente distintas (como "Bota" e "Vestido"), o que exige métricas como Precisão, Recall e F1-Score.Além disso, em cenários onde as classes pudessem estar desbalanceadas, um modelo poderia apresentar uma acurácia alta apenas por classificar corretamente a classe majoritária, falhando completamente em identificar as minoritárias. Portanto, a Matriz de Confusão torna-se indispensável para diagnosticar onde o modelo realmente precisa de melhorias.

**2. Como você garante que o resultado não ocorreu por acaso?**
> Para garantir a robustez estatística e evitar que o resultado seja fruto de uma partição "sortuda" dos dados, a técnica ideal é a Validação Cruzada ($K$-Fold). Em vez de uma única divisão entre treino e teste, o dataset é particionado em $k$ subconjuntos, onde o modelo é treinado e testado $k$ vezes, alternando o conjunto de teste a cada rodada.A média e o desvio padrão das métricas obtidas nessas $k$ execuções fornecem uma estimativa muito mais confiável da capacidade de generalização do modelo. Além disso, o uso de testes de hipótese estatística (como o teste t de Student) poderia ser aplicado para confirmar se a diferença de performance entre o Random Forest e o AdaBoost é estatisticamente significante.

**3. Cite dois possíveis problemas metodológicos neste experimento.**

> O primeiro problema é o Risco de Overfitting no Conjunto de Teste. Como utilizamos o conjunto de teste para tomar decisões sobre quais hiperparâmetros (como o n_estimators) funcionam melhor, acabamos "viciando" nossa escolha final naquelas amostras específicas. O correto seria utilizar um terceiro conjunto, o de Validação, mantendo o de teste intocado até o fim do processo.O segundo problema é a Busca de Hiperparâmetros Limitada. Avaliamos apenas a quantidade de estimadores, ignorando outros parâmetros críticos como a profundidade máxima (max_depth) ou a taxa de aprendizado (learning_rate). Sem uma busca exaustiva (como Grid Search ou Random Search), não podemos afirmar que encontramos o melhor modelo possível para nenhum dos dois algoritmos.

**4. O pipeline implementado é confiável? Justifique.**
> O pipeline é confiável como um baseline experimental e acadêmico, mas não para um ambiente de produção. Ele é sólido no quesito reprodutibilidade, pois o controle rigoroso de sementes (random_state) e a modularização das funções garantem que o experimento possa ser auditado e repetido com os mesmos resultados por qualquer pesquisador.Contudo, sua confiabilidade científica é limitada pela falta de automação na seleção de atributos e pela ausência de uma validação cruzada robusta. Para torná-lo profissional, seria necessário integrar técnicas de pré-processamento mais avançadas e uma camada de otimização de hiperparâmetros que garantisse a estabilidade do modelo frente a diferentes variações do dataset.
